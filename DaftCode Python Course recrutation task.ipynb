Jesteś pracownikiem w firmie spedycyjnej "TEU Logistics". Twoja firma zajmuje się przewozem towarów w morskich kontenerach własnymi statkami po całym świecie. Manager poprosił Cię o małą analizę danych, dlatego wysłał Ci plik z danymi (CSV), w którym są dane aktualnie transportowanych kontenerów przez "TEU Logistics". Dane są trochę mało czytelne i jest ich za dużo na "ręczne" kalkulacje.

W pierwszym wierszu, w każdej kolumnie podany jest numer, nazwa i klasa statku, który przewozi ładunek. W kolejnych wierszach, w każdej kolumnie widnieją dane kontenerów znajdujących się na pokładzie danego statu w kolumnie wraz z kwotą jaka została pobrana za przewóz kontenera.

Szczegóły: Nazwa statku, jego port startowy oraz docelowy i klasa podana jest w formacie:

aa-bb: xxxxxx (ttttttt)
gdzie:

aa - kraj pochodzenia kontenera
bb - kraj docelowy kontenera
xxxxxx: nazwa statku
tttttt: klasa statku
np: ES-PL: Cool Ship (Small Feeder)

Każdy kontener ma identyfikator nadany w formacie:

aa-bb-cccccccc/yyyy/xx@ddddddddd.ee
gdzie:

aa - kraj pochodzenia kontenera
bb - kraj docelowy kontenera
ccccccccc - numer kontenera (nie jest unikalny!!)
yyyy - waga kontenera w kilogramach (cyfrowo 0001-9999)
xx - typ ładunku w kontenerze (A0-Z9)
dddddddd.ee - nazwa i kraj pochodzenia firmy, która nadaje kontener, kraj jest 2 literowym oznaczeniem wg ISO 3166-1
długość pól cccccc, dddddd nie jest stała
pl-jp-2343432/2201/A1@companyname.pl oraz pl-jp-1223123/2201/A1@companyname.de to kontenery 2 różnych firm wg systemu informatycznego Twojej firmy.
Rekord jest połączeniem identyfikatora kontenera i kwoty w formacie:

identyfikator/kwota:
pl-jp-2343432/3100/Z1@companyname.pl/83427
Z wszelakich powodów takich jak różne kontrakty, terminy, kolejność załadunku, waga i rodzaj ładunku, cena wysyłki kontenera jest bardzo różna. Zauważ, że kraj pochodzenia i destynacja kontenera nie jest tożsamy z portem docelowym dla kontenera. Często taki kontenery trzeba przeładować kilka razy.

CSV: Format zapisu danych w którym rekordy oddzielone są od siebie znakiem średnika ';' a wiersze znakiem nowej linii '\n'

Odpowiedziami na zadanie są liczby lub jedno konkretne słowo. Odpowiedzi zawierające kalkulacje, kod, słowne zapisy liczb czy komentarze nie będą brane pod uwagę i będą oceniane na 0 punktów. ## PYTANIE NR. 3 Jaka jest średnia waga kontenera z przetworami owocowymi (A3) z dokładnością do 1 kg np: 1234? (zaokrąglane w górę) – 20 pkt

```python
weight_objects=0
weight_sum=0

for ship_info in df.columns:
    weights = df[ship_info].str.extractall(r'[A-Za-z]{2}-[A-Za-z]{2}-\d+/(.+)/A3')
    weight_objects += weights.shape[0]    
    weights.reset_index(drop=True, inplace=True)
    weights = weights.astype(int)
    weight_sum += weights.sum(axis=0)

avg_weight = weight_sum/weight_objects

import math

math.ceil(avg_weight)
```

## PYTANIE NR. 4 Która Polska firma globalnie przewozi najwięcej kontenerów? (nazwa firmy) – 25 pkt

```python
polish_companys_dict = {}

for ship_info in df.columns:
    polish_companys = df[ship_info].str.extractall(r'[A-Za-z]{2}-[A-Za-z]{2}-\d+/\d{4}/[A-Z0-9]{2}@(.+)\.pl')
    polish_companys.reset_index(drop=True, inplace=True)
    for index, company in polish_companys.iterrows():
        if not company[0] in polish_companys_dict:
            polish_companys_dict[company[0]] = 0 
            
        polish_companys_dict[company[0]] += 1
        
        
best_polish_company = (0, 0)
for key, value in polish_companys_dict.items():
    if best_polish_company[1] < value:
        best_polish_company = (key, value)
        
        
print(best_polish_company)
```

## PYTANIE NR. 5 Jakiego typu ładunek o największej wartości eksportują Niemieckie firmy z Niemiec? (wartość to stosunek ceny do wagi) – 25 pkt
